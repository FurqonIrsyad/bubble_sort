```python
import pandas as pd
import csv

def read_csv(file_name):
    return pd.read_csv(file_name, delimiter=';')

def bubble_sort(df, grade_key):
    data = df.to_dict('records')
    n = len(data)
    for i in range(n):
        for j in range(0, n - i - 1):
            if float(data[j][grade_key]) > float(data[j + 1][grade_key]):
                data[j], data[j + 1] = data[j + 1], data[j]
    return pd.DataFrame(data)

if __name__ == "__main__":
    input_file = '/content/sample_data/student_grade.csv'

    student_data = read_csv(input_file)

    sorted_data = bubble_sort(student_data, 'english.grade')

    print("\nTabel setelah sorting berdasarkan 'english.grade':")
    display(sorted_data)
```
